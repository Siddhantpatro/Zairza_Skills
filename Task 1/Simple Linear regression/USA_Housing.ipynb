```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import seaborn as sns

%matplotlib inline
```

##### Importing the USA housing dataset

```python
housing_data = pd.read_csv('USA_Housing.csv')
housing_data.head()
```

##### Data Cleaning Process

##### Finding, NaN value in the dataset

```python
housing_data.isnull().sum()
```

##### Using LabelEncoder() to convert categorial data to numeric 

```python
le = LabelEncoder()
```

```python
housing_data['Address'] = le.fit_transform(housing_data['Address'])
```

```python
housing_data['Avg. Area Income'] = le.fit_transform(housing_data['Avg. Area Income'])
housing_data['Avg. Area House Age'] = le.fit_transform(housing_data['Avg. Area House Age'])
housing_data['Avg. Area Number of Rooms'] = le.fit_transform(housing_data['Avg. Area Number of Rooms'])
housing_data['Avg. Area Number of Bedrooms'] = le.fit_transform(housing_data['Avg. Area Number of Bedrooms'])
housing_data['Price'] = le.fit_transform(housing_data['Price'])
housing_data['Area Population'] = le.fit_transform(housing_data['Area Population'])
```

```python
housing_data.head()
```

##### Separating target / dependent column as 'y' from the dataset

```python
y = np.array(housing_data['Price'])
y
```

```python
y = y.reshape(-1,1)
```

```python
y.shape
```

##### Column Avg. Area Number of Rooms is considered as 'X' 

```python
X = housing_data['Avg. Area Number of Rooms']
X.head()
```

```python
X = np.array(X).reshape(-1,1)
X.shape
```

##### Using train_test_split method
##### Training size = 60 % and test size = 40 % 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4)
```

```python
len(X_train)
```

```python
len(X_test)
```

##### Standardization of X_train and y_train using StandardScaler 

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

##### Linear Regression algorithm: Fitting of the trained data

```python
linear = LinearRegression()
```

```python
linear.fit(X_train, y_train)
```

```python
linear.intercept_
```

```python
linear.coef_
```

##### Prediction of the 0th row of the 'X' dataset

```python
linear.predict([[2512]])
```

##### y = mx + c

```python
486.63346823*2512 
```

##### Prediction of X_test

```python
linear.predict(X_test)
```

##### Graph between the x and y training set

```python
plt.figure(figsize = (10,6))

plt.scatter(X_train, y_train, color = 'red', label = 'Datapoints')
plt.plot(X_test, linear.predict(X_test), color = 'black', label = 'Linear Regression')

plt.xlim(0,2)
plt.ylim(0,8000)

plt.xlabel('Avg Area Number of Rooms')
plt.ylabel('Price')

plt.grid(True)

plt.title('Graph for Training data')

plt.legend()

plt.style.use('fivethirtyeight')
```

##### Graph between the x and y testing set

```python
plt.figure(figsize = (10,6))

plt.scatter(X_test, y_test, color = 'red', label = 'Datapoints')
plt.plot(X_test, linear.predict(X_test), color = 'black', label = 'Linear Regression')

plt.xlim(0,2)
plt.ylim(0,8000)

plt.xlabel('Avg Area Number of Rooms')
plt.ylabel('Price')

plt.grid(True)

plt.title('Graph for Testing data')

plt.legend()

plt.style.use('fivethirtyeight')
```
